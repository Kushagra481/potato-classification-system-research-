```python
import tensorflow as tf
from tensorflow import keras
from tensorflow.keras import layers
import numpy as np
import matplotlib.pyplot as plt
import os
import glob
from PIL import Image

# Set random seeds for reproducibility
tf.random.set_seed(42)
np.random.seed(42)

# Global class_names list for consistent access
class_names = ['Bacteria', 'Fungi', 'Healthy', 'Nematode', 'Pest', 'Phytophthora', 'Virus']

class ConditionalGAN:
    def __init__(self, img_shape=(64, 64, 3), num_classes=7, latent_dim=100):
        self.img_shape = img_shape
        self.num_classes = num_classes
        self.latent_dim = latent_dim
        
        # Build and compile the discriminator
        self.discriminator = self.build_discriminator()
        self.discriminator.compile(
            loss=['binary_crossentropy', 'sparse_categorical_crossentropy'],
            optimizer=keras.optimizers.Adam(0.0002, 0.5),
            metrics={'validity': 'accuracy', 'label': 'accuracy'}
        )
        
        # Build the generator
        self.generator = self.build_generator()
        
        # For the combined model, we only train the generator
        self.discriminator.trainable = False
        
        # The generator takes noise and label as input
        # and generates images
        noise = layers.Input(shape=(self.latent_dim,))
        label = layers.Input(shape=(1,))
        img = self.generator([noise, label])
        
        # The discriminator takes generated images as input
        # and determines validity and class
        validity, target_label = self.discriminator(img)
        
        # The combined model (stacked generator and discriminator)
        self.combined = keras.Model([noise, label], [validity, target_label])
        self.combined.compile(
            loss=['binary_crossentropy', 'sparse_categorical_crossentropy'],
            optimizer=keras.optimizers.Adam(0.0002, 0.5)
        )
    
    def build_generator(self):
        noise = layers.Input(shape=(self.latent_dim,))
        label = layers.Input(shape=(1,), dtype='int32')
        
        # Embed label to same dimension as noise
        label_embedding = layers.Embedding(self.num_classes, self.latent_dim)(label)
        label_embedding = layers.Flatten()(label_embedding)
        
        # Concatenate noise and label
        model_input = layers.Concatenate()([noise, label_embedding])
        
        # Dense layers
        x = layers.Dense(256)(model_input)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        x = layers.Dense(512)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        x = layers.Dense(1024)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        # Reshape to start of conv layers
        x = layers.Dense(8 * 8 * 256)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.Reshape((8, 8, 256))(x)
        
        # Upsampling layers
        x = layers.Conv2DTranspose(128, (5, 5), strides=(2, 2), padding='same')(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        
        x = layers.Conv2DTranspose(64, (5, 5), strides=(2, 2), padding='same')(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        
        x = layers.Conv2DTranspose(32, (5, 5), strides=(2, 2), padding='same')(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        
        # Final layer
        img = layers.Conv2DTranspose(self.img_shape[2], (5, 5), strides=(1, 1), 
                                   padding='same', activation='tanh')(x)
        
        return keras.Model([noise, label], img, name='generator')
    
    def build_discriminator(self):
        img = layers.Input(shape=self.img_shape)
        
        # Convolutional layers
        x = layers.Conv2D(64, (3, 3), strides=(2, 2), padding='same')(img)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.Dropout(0.25)(x)
        
        x = layers.Conv2D(128, (3, 3), strides=(2, 2), padding='same')(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.Dropout(0.25)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        x = layers.Conv2D(256, (3, 3), strides=(2, 2), padding='same')(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.Dropout(0.25)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        x = layers.Conv2D(512, (3, 3), strides=(2, 2), padding='same')(x)
        x = layers.LeakyReLU(alpha=0.2)(x)
        x = layers.Dropout(0.25)(x)
        x = layers.BatchNormalization(momentum=0.8)(x)
        
        x = layers.Flatten()(x)
        
        # Validity output - NAMED 'validity'
        validity = layers.Dense(1, activation='sigmoid', name='validity')(x)
        
        # Class label output - NAMED 'label'
        label = layers.Dense(self.num_classes, activation='softmax', name='label')(x)
        
        return keras.Model(img, [validity, label], name='discriminator')
    
    def train(self, dataset, epochs=10000, batch_size=32, save_interval=1000):
        # Load and preprocess your potato disease dataset
        X_train, y_train = dataset  # Assuming dataset is (images, labels)
        
        # Robust dataset validation
        if not isinstance(X_train, np.ndarray) or not isinstance(y_train, np.ndarray):
            raise TypeError("Dataset components must be NumPy arrays.")
            
        if X_train.size == 0 or y_train.size == 0:
            raise ValueError("Dataset is empty! Please check your dataset path and loading function.")
        
        # This is the crucial check for the shape's rank
        if X_train.ndim != 4: # Expected shape (num_samples, height, width, channels)
            raise ValueError(f"Expected X_train to have 4 dimensions (N, H, W, C), but got {X_train.ndim} dimensions. Actual shape: {X_train.shape}")
        
        print(f"Training dataset shape: {X_train.shape}")
        print(f"Labels shape: {y_train.shape}")
        print(f"Unique classes: {np.unique(y_train)}")
        
        # Adjust batch size if dataset is smaller
        if X_train.shape[0] < batch_size:
            batch_size = X_train.shape[0]
            print(f"Adjusted batch size to {batch_size} due to small dataset")
            # If batch_size becomes 0, it will cause issues, so handle that
            if batch_size == 0:
                raise ValueError("Calculated batch size is 0. No data available for training.")
        
        # Normalize images to [-1, 1]
        X_train = (X_train.astype(np.float32) - 127.5) / 127.5
        
        # Adversarial ground truths
        valid = np.ones((batch_size, 1))
        fake = np.zeros((batch_size, 1))
        
        for epoch in range(epochs):
            # ---------------------
            #  Train Discriminator
            # ---------------------
            
            # Select a random batch of images
            idx = np.random.randint(0, X_train.shape[0], batch_size)
            imgs = X_train[idx]
            img_labels = y_train[idx]
            
            # Generate fake images
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))
            sampled_labels = np.random.randint(0, self.num_classes, batch_size)
            gen_imgs = self.generator.predict([noise, sampled_labels], verbose=0)
            
            # FIXED: Ensure labels have proper shape for sparse_categorical_crossentropy
            # Real images training
            d_loss_real = self.discriminator.train_on_batch(imgs, [valid, img_labels])
            
            # Fake images training - use sampled_labels directly (not reshaped)
            d_loss_fake = self.discriminator.train_on_batch(gen_imgs, [fake, sampled_labels])
            d_loss = 0.5 * np.add(d_loss_real, d_loss_fake)
            
            # ---------------------
            #  Train Generator
            # ---------------------
            
            noise = np.random.normal(0, 1, (batch_size, self.latent_dim))
            sampled_labels = np.random.randint(0, self.num_classes, batch_size)
            
            # Train generator (wants discriminator to mistake images as real)
            # Use sampled_labels directly for both generator input and target
            g_loss = self.combined.train_on_batch([noise, sampled_labels], [valid, sampled_labels])
            
            # Print progress
            if epoch % 100 == 0:
                print(f"Epoch {epoch}, D total loss: {d_loss[0]:.4f}, D validity acc: {100*d_loss[3]:.2f}%, D label acc: {100*d_loss[4]:.2f}%, G total loss: {g_loss[0]:.4f}")
            
            # Save generated images at intervals
            if epoch % save_interval == 0:
                self.save_imgs(epoch)
    
    def save_imgs(self, epoch):
        r, c = 2, 5
        noise = np.random.normal(0, 1, (r * c, self.latent_dim))
        # Create labels for each class (ensure we have enough)
        sampled_labels = np.arange(0, min(self.num_classes, r * c))
        
        # If we need more labels than classes, repeat some
        if len(sampled_labels) < r * c:
            additional_labels = np.random.randint(0, self.num_classes, r * c - len(sampled_labels))
            sampled_labels = np.concatenate([sampled_labels, additional_labels])
        
        gen_imgs = self.generator.predict([noise, sampled_labels], verbose=0)
        
        # Rescale images to [0, 1] and convert to uint8 for Pillow
        gen_imgs = ((gen_imgs * 0.5 + 0.5) * 255).astype(np.uint8)
        
        # Create a directory if it doesn't exist
        output_dir = '/kaggle/working/training_progress'
        os.makedirs(output_dir, exist_ok=True)
        
        fig, axs = plt.subplots(r, c, figsize=(15, 6))
        cnt = 0
        for i in range(r):
            for j in range(c):
                axs[i,j].imshow(gen_imgs[cnt])
                # Use class_names directly for titles
                axs[i,j].set_title(f'Class {class_names[sampled_labels[cnt]]}')
                axs[i,j].axis('off')
                cnt += 1
        fig.savefig(os.path.join(output_dir, f"generated_images_{epoch}.png"))
        plt.close()
    
    def generate_samples(self, class_label, num_samples=10):
        """Generate samples for a specific class"""
        noise = np.random.normal(0, 1, (num_samples, self.latent_dim))
        labels = np.full(num_samples, class_label)  # Changed from (num_samples, 1) to (num_samples,)
        
        generated_imgs = self.generator.predict([noise, labels], verbose=0)
        # Rescale to [0, 1]
        generated_imgs = 0.5 * generated_imgs + 0.5
        
        return generated_imgs

# Dataset loading for Kaggle Potato Leaf Disease Dataset
def load_potato_dataset(dataset_path):
    """
    Load the Kaggle Potato Leaf Disease Dataset
    Classes: 0-Bacteria, 1-Fungi, 2-Healthy, 3-Nematode, 4-Pest, 5-Phytophthora, 6-Virus
    """
    from tensorflow.keras.preprocessing.image import load_img, img_to_array
    
    # Class mapping - CORRECTED TO MATCH CAPITALIZATION
    class_names_local = ['Bacteria', 'Fungi', 'Healthy', 'Nematode', 'Pest', 'Phytophthora', 'Virus']
    class_to_idx = {name: idx for idx, name in enumerate(class_names_local)}
    
    images = []
    labels = []
    
    print(f"Attempting to load dataset from: {dataset_path}")
    
    # Debug: Check if base path exists
    if not os.path.exists(dataset_path):
        print(f"Dataset path does not exist: {dataset_path}")
        return None, None
    
    # Load images from each class folder
    for class_name_folder in class_names_local:
        class_path = os.path.join(dataset_path, class_name_folder)
        
        if os.path.exists(class_path):
            print(f"Loading {class_name_folder} images from {class_path}...")
            
            # Get all image files (jpg, png, jpeg)
            image_patterns = [
                os.path.join(class_path, "*.jpg"),
                os.path.join(class_path, "*.JPG"),
                os.path.join(class_path, "*.png"),
                os.path.join(class_path, "*.PNG"),
                os.path.join(class_path, "*.jpeg"),
                os.path.join(class_path, "*.JPEG")
            ]
            
            image_files = []
            for pattern in image_patterns:
                image_files.extend(glob.glob(pattern))
            
            print(f"Found {len(image_files)} image files in {class_name_folder}")
            
            if len(image_files) == 0:
                print(f"No images found in {class_path}. Skipping.")
                continue
            
            loaded_count = 0
            for img_path in image_files:
                try:
                    # Load and resize image
                    img = load_img(img_path, target_size=(64, 64))
                    img_array = img_to_array(img)
                    images.append(img_array)
                    labels.append(class_to_idx[class_name_folder])
                    loaded_count += 1
                except Exception as e:
                    print(f"Error loading {img_path}: {e}")
                    continue
            
            print(f"Successfully loaded {loaded_count} images from {class_name_folder}")
        else:
            print(f"Class path does not exist: {class_path}. Skipping.")
    
    if len(images) == 0:
        print("No images were loaded! Returning None, None.")
        return None, None
    
    X_train = np.array(images)
    y_train = np.array(labels)
    
    # Ensure X_train has 4 dimensions (N, H, W, C) even if only one image is loaded
    if X_train.ndim == 3 and X_train.shape[0] > 0 and X_train.shape[-1] in [1, 3]:
        print("Warning: X_train has 3 dimensions. Reshaping to (1, H, W, C).")
        X_train = np.expand_dims(X_train, axis=0)
    elif X_train.ndim != 4:
        print(f"Warning: X_train has {X_train.ndim} dimensions after loading. Expected 4. Shape: {X_train.shape}")
    
    print(f"\nDataset loaded successfully!")
    print(f"Total images: {X_train.shape[0]}")
    print(f"Image shape: {X_train.shape[1:]}")
    print(f"Classes: {len(class_names_local)}")
    print(f"Class distribution:")
    unique, counts = np.unique(y_train, return_counts=True)
    for class_idx, count in zip(unique, counts):
        print(f"  {class_names_local[class_idx]}: {count} images")
    
    return X_train, y_train

# Complete training script for Kaggle environment
def train_potato_disease_gan():
    """
    Complete training pipeline for the Kaggle Potato Disease Dataset
    """
    # Set up directories
    os.makedirs('/kaggle/working/generated_samples', exist_ok=True)
    os.makedirs('/kaggle/working/training_progress', exist_ok=True)
    
    # Load dataset
    print("Loading Kaggle Potato Disease Dataset...")
    
    # Primary dataset path
    dataset_path_primary = '/kaggle/input/potato-leaf-disease-dataset/Potato Leaf Disease Dataset in Uncontrolled Environment'
    dataset = load_potato_dataset(dataset_path_primary)
    
    # If primary path failed, try alternative paths
    if dataset[0] is None:
        print("\nPrimary dataset path failed. Trying alternative paths...")
        alternative_paths = [
            '/kaggle/input/potato-leaf-disease-dataset',
            '/kaggle/input/potato-disease-dataset',
        ]
        
        for alt_path in alternative_paths:
            print(f"Trying alternative path: {alt_path}")
            if os.path.exists(alt_path):
                dataset = load_potato_dataset(alt_path)
                if dataset[0] is not None:
                    print(f"Successfully loaded dataset from: {alt_path}")
                    break
        
        if dataset[0] is None:
            print("Could not load dataset from any known path! Please verify the dataset's location.")
            return None, None
    
    # Create GAN
    print("\nCreating Conditional GAN...")
    gan = ConditionalGAN(img_shape=(64, 64, 3), num_classes=len(class_names), latent_dim=100)
    
    # Print model summaries
    print("\nGenerator Summary:")
    gan.generator.summary()
    print("\nDiscriminator Summary:")
    gan.discriminator.summary()
    
    # Train the GAN
    print("\nStarting GAN training...")
    gan.train(dataset, epochs=5000, batch_size=16, save_interval=500)
    
    # Save the trained models
    gan.generator.save('/kaggle/working/potato_disease_generator.h5')
    gan.discriminator.save('/kaggle/working/potato_disease_discriminator.h5')
    print("Trained models saved successfully.")

    # Generate and save samples for each class
    print("\nGenerating samples for each disease class...")
    for class_id, class_name_str in enumerate(class_names):
        samples = gan.generate_samples(class_id, num_samples=20)
        
        # Save individual samples
        for i, sample in enumerate(samples):
            sample_img = (sample * 255).astype(np.uint8)
            img_pil = Image.fromarray(sample_img)
            img_pil.save(f'/kaggle/working/generated_samples/{class_name_str.replace(" ", "_")}_sample_{i+1}.png')
        
        # Create comparison grid
        fig, axes = plt.subplots(4, 5, figsize=(20, 16))
        fig.suptitle(f'Generated {class_name_str} Samples', fontsize=16)
        
        for i, img in enumerate(samples):
            row = i // 5
            col = i % 5
            axes[row, col].imshow(img)
            axes[row, col].set_title(f'Sample {i+1}')
            axes[row, col].axis('off')
        
        plt.tight_layout()
        plt.savefig(f'/kaggle/working/{class_name_str.replace(" ", "_")}_generated_grid.png', dpi=150, bbox_inches='tight')
        plt.close()
    print("Generated samples saved.")

    # Create augmented dataset
    print("\nCreating augmented dataset...")
    X_augmented, y_augmented = augment_dataset_with_gan(dataset, gan, target_samples_per_class=500)
    
    # Save augmented dataset info
    if X_augmented is not None:
        print("Augmented dataset statistics:")
        unique, counts = np.unique(y_augmented, return_counts=True)
        with open('/kaggle/working/augmented_dataset_stats.txt', 'w') as f:
            f.write("Augmented Dataset Statistics:\n")
            f.write(f"Total samples: {len(y_augmented)}\n")
            f.write("Class distribution:\n")
            for class_idx, count in zip(unique, counts):
                f.write(f"  {class_names[class_idx]}: {count} samples\n")
        print("Augmented dataset statistics saved.")
    else:
        print("Augmentation skipped due to issues with original dataset.")

    print("\nTraining completed successfully!")
    print("Check /kaggle/working/ directory for:")
    print("- Generated samples")
    print("- Trained models")
    print("- Training progress images")
    print("- Augmented dataset statistics")
    
    return gan, (X_augmented, y_augmented)

# Data augmentation helper
def augment_dataset_with_gan(original_dataset, gan_model, target_samples_per_class=1000):
    """
    Augment your original dataset with GAN-generated samples
    """
    X_original, y_original = original_dataset
    
    if X_original is None or len(X_original) == 0:
        print("Cannot augment empty dataset!")
        return None, None
    
    # Count samples per class
    unique_classes, counts = np.unique(y_original, return_counts=True)
    
    X_augmented_list = []
    y_augmented_list = []
    
    # Add original data to the lists
    X_augmented_list.append(X_original)
    y_augmented_list.append(y_original)

    for class_id, count in zip(unique_classes, counts):
        if count < target_samples_per_class:
            # Generate additional samples for this class
            needed_samples = target_samples_per_class - count
            print(f"Generating {needed_samples} samples for class {class_names[class_id]}...")
            generated_samples = gan_model.generate_samples(class_id, needed_samples)
            
            # Convert back to [0, 255] range for consistency
            generated_samples = (generated_samples * 255).astype(np.uint8)
            
            X_augmented_list.append(generated_samples)
            y_augmented_list.append(np.full(needed_samples, class_id))
        else:
            print(f"Class {class_names[class_id]} already has {count} samples, no augmentation needed.")
    
    X_combined = np.vstack(X_augmented_list)
    y_combined = np.hstack(y_augmented_list)
    
    return X_combined, y_combined

# Quick test function
def test_gan_generation():
    """
    Quick test to generate a few samples
    """
    # Load a pre-trained generator if available
    try:
        generator = keras.models.load_model('/kaggle/working/potato_disease_generator.h5', compile=False)
        print("Loaded pre-trained generator")
    except Exception as e:
        print(f"No pre-trained model found or error loading: {e}. Training a new one...")
        gan, _ = train_potato_disease_gan()
        if gan is None:
            print("Failed to train GAN, cannot perform generation test.")
            return
        generator = gan.generator
    
    # Generate test samples
    fig, axes = plt.subplots(1, 7, figsize=(21, 3))
    for i in range(7):
        noise = np.random.normal(0, 1, (1, 100))
        label = np.array([i])  # Changed from [[i]] to [i]
        
        sample = generator.predict([noise, label], verbose=0)[0]
        sample = 0.5 * sample + 0.5  # Rescale to [0, 1] for display
        
        axes[i].imshow(sample)
        axes[i].set_title(f'{class_names[i]}')
        axes[i].axis('off')
    
    plt.tight_layout()
    plt.savefig('/kaggle/working/test_generation.png', dpi=150, bbox_inches='tight')
    plt.show()

# Debug function to explore dataset structure
def debug_dataset_structure():
    """
    Debug function to explore the dataset structure
    """
    base_path = '/kaggle/input/'
    
    print("Exploring Kaggle input directory structure...")
    
    try:
        print("\nContents of /kaggle/input/:")
        for item in os.listdir(base_path):
            item_path = os.path.join(base_path, item)
            print(f"  {item} ({'directory' if os.path.isdir(item_path) else 'file'})")
            
            if os.path.isdir(item_path) and 'potato' in item.lower():
                print(f"    Contents of {item}:")
                try:
                    for subitem in os.listdir(item_path):
                        subitem_path = os.path.join(item_path, subitem)
                        print(f"      {subitem} ({'directory' if os.path.isdir(subitem_path) else 'file'})")
                        
                        if os.path.isdir(subitem_path):
                            print(f"        Contents of {subitem}:")
                            try:
                                for subsubitem in os.listdir(subitem_path):
                                    print(f"          {subsubitem}")
                            except Exception as e:
                                print(f"          Error: {e}")
                except Exception as e:
                    print(f"    Error: {e}")
    except Exception as e:
        print(f"Error exploring directory: {e}")

# Run the training
if __name__ == "__main__":
    # You can uncomment debug_dataset_structure() if you're still facing path issues
    # debug_dataset_structure() 
    
    gan, augmented_dataset = train_potato_disease_gan()
    
    if gan is not None:
        print("\nRunning a quick generation test with the trained GAN...")
        test_gan_generation()
    else:
        print("\nGAN training failed, skipping generation test.")
```
